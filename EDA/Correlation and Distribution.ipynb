# Correlation And Distributions
- 상관성과 분포에 대해 파악합니다.

### 패키지 로딩

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 데이터 로딩

```python
train = pd.read_csv('C:/LG_Aimers/data/train.csv')
train = train.iloc[:,1:57]
```

### 상관성 파악

```python
### 상관행렬 불러오기
corr_data = train.corr()

### 상관 행렬에서 상관관계가 0.5 이상인 칼럼들을 딕셔너리 형태로 불러옴
corr_dict = dict()
for i in corr_data.columns :
    corr_dict[i] = list(corr_data[np.abs(corr_data[i])>0.5][i].index)
```

```python
### 결과 보기
corr_dict
```

### 분포파악

```python
### 반북문으로 돌면서 모든 변수에 대한 히스토그림을 생성함.
f,axes = plt.subplots(nrows=56,ncols=1,figsize=(20,300))
for index,value in enumerate(train.columns) :
    plt.subplot(56,1,index+1)
    plt.hist(train[value],bins = 50)
    axes[index].set_title(value,fontsize = 12)
```
